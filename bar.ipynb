```python
#数値計算ライブラリ
import numpy as np
import pandas as pd

#可視化ライブラリ
import matplotlib.pyplot as plt
import japanize_matplotlib
import seaborn as sns

#sklearnの標準化モジュールをインポート
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import FactorAnalysis as FA
from sklearn.decomposition import PCA

from scipy import stats
```

```python
#Excel読み込み
df = pd.read_excel("data1.xlsx")
print(df)
```

```python
# 条件名（A～W）
conditions = [chr(65 + i) for i in range(23)]  # 65 は 'A' のASCIIコード

# グラフの作成（23条件 × 4択 の棒グラフ）
ax = df.plot(kind='bar', figsize=(14, 6), color=['skyblue', 'orange', 'red', 'green'], width=0.8)

# タイトルとラベル
plt.xlabel('Condition')
plt.ylabel('Number of Respondents')
# plt.legend(title="Options")

# X軸のラベルを"A"～"W"に設定
plt.xticks(ticks=range(len(conditions)), labels=conditions, rotation=0)  # ラベルを明示的に設定

# 凡例をグラフの右外に配置
ax.legend(loc='upper left', bbox_to_anchor=(1, 1))

# グリッド表示
plt.grid(axis='y', linestyle='--', alpha=0.7)

# グラフを表示
plt.show()
```
