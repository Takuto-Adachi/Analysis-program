```python
import numpy as np

def Bezier(p, t):
    *_, n = p.shape
    J = np.vectorize(lambda i, t: (n/np.r_[1:i+1]-1).prod() * t**i * (1-t)**(n-1-i), signature='(),(m)->(m)')
    return p @ J(np.r_[:n], t)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import pyplot
from matplotlib.offsetbox import OffsetImage, AnnotationBbox

plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["font.size"] = 20

x = np.arange(-100,101)
y = x**2

plt.plot(x+100,-y/100+100, label="Reality trajectory")
plt.plot(x+100,(-y/100+100)*2, label="Parabolic")


plt.tick_params(labelbottom=False, labelleft=False)
plt.tick_params(bottom=False, left=False)

plt.legend(frameon = False, loc="center left", bbox_to_anchor = (1,0.5))

plt.ylim(0,210)

plt.xlabel("Time")
plt.ylabel("Height")

plt.savefig('VRTrajectory.svg', bbox_inches="tight", transparent = True)
```
